Preprocessing for Federated Intrusion Detection Evaluation


```python
import numpy
import statistics
import os

res = []
all_accs = []
k = 0


for filename in os.listdir(os.getcwd()):
    print("Reading " + filename + "...")
    name, file_extension = os.path.splitext(filename)
    if 'client' in name:
        acc_client = []
        f=open(filename,"r")
        lines=f.readlines()
        for x in lines:
            acc_client.append(float(x.split(',')[3].replace("\n", "")))
        f.close()
        #print(acc_client)
        all_accs.append(acc_client)

def deviation_candlesticks():
    for i in range(0,len(all_accs[0])):
        if i % 10 == 0:
            current_val = [all_accs[0][i], all_accs[1][i], all_accs[2][i], all_accs[3][i], all_accs[4][i], all_accs[5][i], all_accs[6][i], all_accs[7][i], all_accs[8][i], all_accs[9][i]]
            print(f"{i}	  {round((min(current_val))*100,2)}	  {round((max(current_val))*100,2)}	  {round((numpy.mean(current_val)-statistics.stdev(current_val))*100,2)}	  {round((numpy.mean(current_val)+statistics.stdev(current_val))*100,2)}")

for i in range(0,len(all_accs[0])):
    #if i %10 ==0:
    current_val = [all_accs[0][i], all_accs[1][i], all_accs[2][i], all_accs[3][i], all_accs[4][i], all_accs[5][i], all_accs[6][i], all_accs[7][i], all_accs[8][i], all_accs[9][i]]

    for k in current_val:
        print(round(k,2), ",", end="")# 	 
    print("")





```
